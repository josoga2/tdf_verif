```python
counter = 0
lister = []

for i,j in zip(tts['TTS start'], tts['TTS end']):
    m = range(i,j)
    if 50 in list(m):
        lister.append(1)
        counter = counter+1
    else:
        lister.append(0)
print (counter)
print(len(lister))
```

```python
listArray = np.array(lister)
tts["range"] = listArray
```
